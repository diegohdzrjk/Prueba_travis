```julia
function randHermitian(n)
    a = rand(n,n)+1im*rand(n,n)
    b = triu(a,1)+conj(transpose(triu(a,1)))+Diagonal(rand(n,n))
    return sort(eigvals(b))
end

function dist_eigvals(n, N)
    dist = zeros((N-1)*n)
    for i in 0:(n-1)
        eigenvals = randHermitian(N)
        for j in 2:N
            dist[i*(N-1)+j-1] = abs(eigenvals[j]-eigenvals[j-1])
        end
    end
    return dist
end
```

```julia
fig = figure("pyplot_histogram",figsize=(10,10))
ax = axes()
grid("on")
xlabel("X")
ylabel("Y")
title("Histogram")

plt[:hist](dist_eigvals(100,40), 1000);
```
